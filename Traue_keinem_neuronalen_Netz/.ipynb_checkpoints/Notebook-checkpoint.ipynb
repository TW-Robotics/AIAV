## Worum gehts?
Sie haben bereits Convolutional Neuronal Networks (CNNs) kennengelernt (z.B.: [hier](https://youtu.be/xIXGDepixcg) auf unserer Plattform). Bisher haben wir CNNs als Framework für Klassifizierung von Bilder und [Lokalisierung von Objekten](https://www.aiav.technikum-wien.at/ai-anwenden) kennengelernt.

Wir beschäftigen uns in diesem Tutorial mit der Frage **Was hat das CNN tatsächlich gelernt?**. Es muss Gründe geben, warum der Output des CNN eine bestimmte Klasse für wahrscheinlicher hält. Im Vergleich zu CNN Anwendungen und Weiterentwicklung existiert hier relativ wenig Forschung. Eine Technik zur Visualisierung der Entscheidungsgründe ist [LRP](https://journals.plos.org/plosone/article?id=10.1371/journal.pone.0130140) oder [GradCAM](https://arxiv.org/abs/1610.02391). Mit diesen Techniken werden wir uns in diesem Tutorial beschäftigen.

Ein weitaus schlimmeres grundsätzliches und häufig nicht sichtbares Problem basiert auf den Daten, die zum Training des CNNs verwendet werden. Vereinfacht kann man dieses Phänomen mit folgendem Beispiel erläutern:

Lernt das CNN nun Pferde zu klassifizieren oder Wasserzeichen? Wie kann man das herausfinden? Dieses Problem ist intensiv in [dieser Publikation](https://www.nature.com/articles/s41467-019-08987-4.pdf) diskutiert.

In diesem Tutorial geht es um ähnliche Dinge: wir wollen unterschiedliche Kleidungsstücke klassifizieren und anschließend feststellen ob die Gründe dieser Klassifizierung valide sind! Dazu implementieren wir folgendes Vorgehen:

1. Datensatz erzeugen
2. CNN trainieren
3. CNN Performance analysieren

Das weitere Vorgehen dieses Tutorials diskutiert dieses Vorgehen.

## Grundlegende Python ToDos und laden der Daten
Für die Verwendung dieses Tutorials können Sie entweder dieses Notebook nutzen oder das dazu gehörende Skript verwenden und anpassen ([siehe AIAV Software](https://AIAVLink)).
Folgende Codezeilen müssen zunächst implementiert werden.

```python
from keras.applications import VGG16          #Wir nutzen Keras und verwenden das VGG16 Basismodell
from keras import models, layers, optimizers  #Einige Keras module
from keras.preprocessing.image import ImageDataGenerator        #Wir nutzen den Keras Datengenerator
from keras.layers import Input, Flatten, Dense, Dropout  #Einige Layer, die wir benötigen
from keras.models import Model #Damit 'bauen' wir uns ein CNN für unsere Anwendung
from sklearn.metrics import confusion_matrix        #Mit den 'confusion matrix' analysieren wir grob die CNN performance
import sys #Für einige Systemfunktionen
import numpy as np #Für Matrixmanipulation
#--- Globale Definitionen ---#
train_dir="./TRAIN"         #Pfad zu unseren Trainingsdaten
validation_dir="./TEST"     #Pfad zu unseren Testdaten
image_size=224              #Wir skalieren alle Bilder auf 244x244 Pixel
train_batchsize = 100       #Die Trainings batch size - also wie viele Bilder wir auf einmal nutzen 
val_batchsize = 100         #Die Validierungs batch size
epochs=50                   #Anzahl an Epochen um das CNN zu trainieren
Augmention=True             #Flag um Augmention zu (de) aktivieren (wir wollen mit Augmention arbeiten)
```

## CNN fine-tuning
Wir nutzen das [VGG-16 CNN](https://arxiv.org/abs/1409.1556) vortrainiert auf [imagenet](https://arxiv.org/pdf/1409.0575.pdf). Dieses vortrainierte Modell (Achtung: das automatische Herunterladen des vortrainierten Models kann dauern) werden wir dann auf unsere Daten anpassen.

```python
vgg_conv = VGG16(weights='imagenet', #Wir laden VGG16 und nutzen die vortrainierten Gewichtungen
                 include_top=False,  #Wir wollen das Netzwerk an unsere Daten anpassen -> das 'top' benötigen wir nicht
                 input_shape=(image_size, image_size, 3) #Alle Bilder werden auf 224x224x3 Pixel skaliert
                )
for layer in vgg_conv.layers[:-4]: #Diese Layer sollen nicht weiter trainiert werden
    layer.trainable = False #Training deaktivieren
#--- Jetzt bauen wir uns das Netzwerk für unsere Zwecke um ---#
last = vgg_conv.output #Wir nehmen den abgeschnittenen Output des VGG16
x = Flatten()(last)                 #... und machen einen Vektor aus dem output
x= Dense(1024,activation='relu')(x) #... und setzen einen fully connected Layer darauf
x=Dropout(0.5)(x)                   #... und fügen einen Dropout Layer dazu
pred=Dense(10,activation='softmax')(x) #... und definieren einen neuen Output Layer
#Anmerkung: 10 Neuronen am Output weil wir 10 Klassen haben
model = Model(vgg_conv.input,pred)  #Wir 'kleben' das Modell zusammen
#---  Show a summary of the model ---#
model.summary() #Damit sehen wir die fertige Architektur
```

## Daten vorbereiten
Als nächsten Schritt müssen wir definieren, wo die Daten fürs Training/Validierung abgelegt wurden und wie wir diese laden wollen.

```python
if(Augmention): #Wenn wir data augmention nutzen wollen erzeugen wir uns einen Data Gernerator, der das unterstützt
    train_datagen = ImageDataGenerator(
        rescale=1./255,
          rotation_range=20,
          width_shift_range=0.2,
          height_shift_range=0.2,
          horizontal_flip=True,
          fill_mode='nearest')
else: #Wenn nicht erstellen wir einen default Data Generator
    train_datagen = ImageDataGenerator(rescale=1./255)
validation_datagen = ImageDataGenerator(rescale=1./255) #Für die Validierung benötigen wir keine Data augmention
#--- Keras Datenstrukturen erzeugen ---#
train_generator = train_datagen.flow_from_directory(
        train_dir, #Pfad zu den Trainingsdaten
        target_size=(image_size, image_size), #Bildgröße
        batch_size=train_batchsize, #Anzahl der Bilder pro Batch
        class_mode='categorical') #Wir wollen Bilder klassifizieren -> also liegen die Daten in Kategorieren vor
#--- Für den Validierung nehmen wir die gleichen Einstellungen ---#
validation_generator = validation_datagen.flow_from_directory(
        validation_dir,
        target_size=(image_size, image_size),
        batch_size=val_batchsize,
        class_mode='categorical',
        shuffle=False #Das zufällige Laden ist hier deaktiviert
)
```

## Trainieren des CNN
Endlich können wir das CNN trainieren. Dies geschieht mit folgenden Codezeilen.

```python
model.compile(loss='categorical_crossentropy', #Wir nutzen die categorical crossentropy als Optimierungsfunktion
             optimizer=optimizers.RMSprop(lr=1e-4), #Diesen Optimizer wollen wir benutzen
             metrics=['acc']) #Die Metrik, die wir zunäcsht Betrachten wollen ist die accuracy
history = model.fit_generator( #Das training startet...
      train_generator, #Unser Data gernerator für Trainignsdaten
      steps_per_epoch=train_generator.samples/train_generator.batch_size, #Umrechnugn in steps pro Epoche
      epochs=epochs, #Vom User definierte Anzahl an Trainingsschritte
      validation_data=validation_generator, #Der definierte Daten Generator für die Validierung
      validation_steps=validation_generator.samples/validation_generator.batch_size, #Siehe oben
      verbose=1 #1: wir wollen den Output sehen; 0:Output deaktiviert
)
#-------------------#
#--- Store model ---#
#-------------------#
print('Validierungs Accuracy:',history.history['val_acc'][-1])   #Plotten der Accuracy des Models
```

## Testen des CNNs
Ähnlich wie bereits erklärt nutzen wir einen Data Generator, um den Testdatensatz zu analysieren. Anschließend sehen wir uns die Confusion Matrix an. Diese zeigt welche Beispiele wie erkannt worden sind.


```python
#--- Erstellen eines Data Generators wie oben beschrieben ---#
train_im = ImageDataGenerator(rescale=1./255)
def test_images():
    train_generator = train_im.flow_from_directory (
            validation_dir, 
            target_size=(224, 224),
            color_mode='rgb',
            batch_size=10000, #Der komplette Datensatz soll auf einmal getestet werden
            shuffle = False,
            class_mode='categorical'
         )
    x =  train_generator
    return x[0][0], x[0][1], train_generator.filenames
test_data = test_images()
#--- Jetzt nutzen wir das CNN und analysieren ---#
prediction = model.predict(test_data[0])
prediction_row = np.argmax(prediction,1)
groundTruth_row = np.argmax(test_data[1],1)
CM = confusion_matrix(prediction_row, groundTruth_row)
print(CM) #Wir plotten die Confusion Matrix
```

Eine ideale Confusion Matrix hätte ausschließlich Einträge in der Diagonalen der Matrix. 

Jetzt können wir uns mit der eigentlichen Frage dieses Tutorials beschäftigen: warum wurden die Objekte wie klassifiziert.
Dazu müssen wir erneut  dazu gehörigen Python Module zunächst laden und anschließend das CNN in eine andere Form bringen.

```python
#--- Importieren der benötigten Module ---#
from vis.utils import utils
from keras import activations
import os
from vis.utils import utils
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.cm as cm
from vis.visualization import visualize_cam, overlay
import innvestigate
import innvestigate.utils as iutils
#--- Aufarbeiten des CNNs, damit die Visualisierungstools mit dem Model arbeiten können ---#
layer_idx = -1 #ID des letzten Layers
model.layers[layer_idx].activation = activations.linear #Adaption der Aktivierungsfunktion
model = utils.apply_modifications(model) #Änderungen annehmen
```

Nun können wir LRP und gradCAM nutzen. Dazu analysieren wir ein Beispiel jeder Klasse

```python
model_wo_softmax = model #Model in eine neue Variable schreiben
LPR_10 = innvestigate.analyzer.relevance_based.relevance_analyzer.LRPAlphaBeta(model_wo_softmax,1,0) #LRP implementierung
ground_truth = test_data[1] #In dieser Variable stehen die wahren Klassenlabels 
labels_ground_truth = np.argmax(ground_truth,1) #Berechnung eines target Vektor - nun sind die Eitnräge die Klassen
for class_ID in range(0,10):
    test_index = np.where(labels_ground_truth==class_ID)[0][0] #Das ist der Index des ersten Testbildes mit ausgewähltem Index
    #--- gradCAM Berechnung ---#
    grads = visualize_cam(model, layer_idx, filter_indices=class_ID, seed_input=test_data[0][test_index], backprop_modifier='guided')
    #--- LRP Berechnung ---#
    LRP = LPR_10.analyze(test_data[0][test_index].reshape((1,224,224,3)))
    #--- Prüfen ob das Objekt richtig erkannt wurde ---#
    if(ground_truth[test_index,class_ID] != 1):
        postfix = " (falsch erkannt)"
    else:
        postfix = ""
    #--- Visualisierung der Daten ---#
    fig, (ax1, ax2, ax3) = plt.subplots(1, 3)
    fig.suptitle("Visualisierung der Klasse "+str(class_ID) + postfix)
    ax1.imshow(test_data[0][test_index]); ax1.axis('off')
    ax2.imshow(grads); ax2.axis('off')
    ax3.imshow(LRP[0,:,:,0]); ax3.axis('off')
```

Die Abbildungen zeigen immer das Rohbild (links), die gradCAM Visualisierung (mittig) und die LRP Visualisierung (rechts). Sehen wir uns die Klassen nacheinander an:

- **Klasse 0**: LRP sieht akzeptabel aus. In der gradCAM Visualisierung erkennen wir, dass der Halsausschnitt genutzt wurde. Andererseits ist der grelle Bereich an der Unterseite wenig nachvollziehbar.
- **Klasse 1**: Sieht nachvollziehbar aus! Anzumerken ist hierbei die gradCAM Visualisierung zwischen den Hosenbeinen. Dies könnte bedeuten, dass schwarze Pixel in dem Bereich (also kein "Stoff") dazu führen, dass Hosen erkannt werden. Das wäre eine sinnvolle Interpretation.
- **Klasse 2**: Hier ist der erkennbare Aufdruck in LRP und GradCAM verdächtig. Dieser Aufdruck sollte nicht repräsentativ für eine Klasse sein.
- **Klasse 3**: GradCAM sieht wenig nachvollziehbar aus. LRP sieht plausibel aus
- **Klasse 4**: LRP ist nachvollziehbar. Die gradCAM Visualisierung zeigt einen eindeutigen Bereich am Objekt, welcher eigentlich nicht für die Klasse repräsentativ ist
- **Klasse 5**: Sieht nachvollziehbar aus
- **Klasse 6-7**: Sieht verdächtig aus
- **Klasse 8**: Bemerken sie den Fokus auf den Bereich des Gurts
- **Klasse 9**: Sieht sehr ähnlich aus wie die bereits betrachteten Schuhe

Der Gesamteindruck zeigt allerdings, dass die nicht alle Entscheidungen des CNN nachvollziehbar sind. 

Wir haben hier mit einem sehr simplem Datensatz gearbeitet - bei komplexeren Datensätzen oder Datensätzen mit wesentlich weniger Beispielen kann diese Visualisierungstechnik wertvollen Einblick in das Verhalten eines CNNs geben.
